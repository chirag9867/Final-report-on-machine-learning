```python
from gensim.models.fasttext import FastText
import nltk
import pandas as pd

w = pd.read_csv('finalcsv.csv')
v = []

for i in range(0, len(w['Lyrics'])):
  v.append(w['Lyrics'][i].split())

ft = FastText(v, vector_size = 300, window = 2, min_count = 1)
```

```python
c = []

for i in range(0, len(w['Lyrics'])):
  c.append(ft.wv[w['Lyrics'][i].split()])

w['Lyrics_vectorized_mean'] = c
```

```python
w.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Unnamed: 0.1.1', 'Album/Label', 'Movie', 'Starring', 'Movie Directors', 'Album', 'Lyrics'], inplace=True, axis = 1)
```

```python
import numpy as np
for i in range(0, len(w.Lyrics_vectorized_mean)):
    w.Lyrics_vectorized_mean[i] = np.mean(w.Lyrics_vectorized_mean[i], axis = 0)
```

```python
singers = pd.get_dummies(w.Singers)
lyricist = pd.get_dummies(w.Lyricist)
composer = pd.get_dummies(w.Composer)
music_dir = pd.get_dummies(w['Music Director'])
music_label = pd.get_dummies(w['Music Label'])
w = pd.concat([w, singers, lyricist, composer, music_dir, music_label], axis = 1)
w.drop(['Singers', 'Lyricist', 'Composer', 'Music Director', 'Music Label'], inplace=True, axis = 1)
```
