```python
import numpy as np
from sklearn import metrics
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split

X = w.drop('new_genre', axis = 1)
Y = w['new_genre']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33)

classifier = MLPClassifier(hidden_layer_sizes=5, max_iter=500, activation='relu', solver='adam')
classifier.fit(X_train, Y_train)

prediction = classifier.predict(X_test)
predictions = [value for value in prediction]
score = np.round(metrics.accuracy_score(Y_test, predictions), 2)
print("Accuracy: %.2f%%" % (score * 100.0))
print(classification_report(Y_test, prediction))
```
