```python
d = []

for i in range(0, len(w['Lyrics'])):
    c = []
    for j in w.Lyrics[i].split():
        c.append(model[j])
    d.append(c)

w['Lyrics_vectorized_mean'] = d
```

```python
w.drop(['Lyrics'], inplace = True, axis = 1)
```

```python
w
```

```python
import numpy as np
for i in range(0, len(w.Lyrics_vectorized_mean)):
    w.Lyrics_vectorized_mean[i] = np.mean(w.Lyrics_vectorized_mean[i], axis = 0)
```

```python
df = pd.DataFrame(w['Lyrics_vectorized_mean'].to_list())
w = pd.concat([w, df], axis = 1)
```

```python
# First XGBoost model for Pima Indians dataset
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# split data into X and y
X = w.drop('Genre', axis = 1)
Y = w['Genre']
# split data into train and test sets
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=test_size)
# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [value for value in y_pred]
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```
